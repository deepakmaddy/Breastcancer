```
import pandas as pd
from google.colab import files
file = files.upload()
X_train = pd.read_csv("xtrain.csv" , header = None)
Y_train = pd.read_csv("ytrain.csv" , header = None)
X_test = pd.read_csv("xtest.csv" , header = None)
Y_test = pd.read_csv("ytest.csv" , header = None)
```

Now I am going to use keras to define a Neural Network that will trained of this data .Thid Neual network can be used to predict value of breast cancer

```
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential() 
#Intializing the ANN
classifier.add(Dense(units=16 , activation = "relu" , input_dim = 30))
classifier.add(Dense(units=8 , activation = "relu" ))
classifier.add(Dense(units=6, activation= "relu" ))
classifier.add(Dense(units=1, activation = "sigmoid"))

```

After that i am going specify loss function and optimization 

```
classifier.compile(optimizer = "rmsprop" , loss = "binary_crossentropy")
```

You now train thwe neural network using classifier.fit passing the training data -i.e the set of X ,this is what the Y should look like .The NN will then spot the pattern in the data ,and bulid a neural network that could replicate that


```
classifier.fit(X_train , Y_train , batch_size = 1 , epochs = 100)
```
